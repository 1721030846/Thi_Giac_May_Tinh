```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# Doc anh
image_path = 'C:/Users/Admin/Desktop/Thi Giac May Tinh/img/acheron.png'
image = Image.open(image_path)

# Anh xam
gray_image = image.convert('L')

# Chuyen anh xam ve mang
image_a = np.array(gray_image)

# Tinh gradient theo huong x (dao ham rieng theo x) bang mat na Prewitt x
gradient_x = ndimage.convolve(image_a, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

# Tinh gradient theo huong y (dao ham rieng theo y) bang mat na Prewitt y
gradient_y = ndimage.convolve(image_a, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Tinh bien do gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Show anh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('nh xm gc')

plt.subplot(2,1,2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('nh xm gc')

plt.show()
```
