```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# duong dan anh
my_path ='C:/Users/Admin/Desktop/Thi Giac May Tinh/img/acheron.png'
# doc anh va hien thi
im = Image.open(my_path)
# display(im)
plt.imshow(im)
```

```python
# chuyen tu anh RGB sang anh xam
im_gray = im.convert("L")
plt.imshow(im_gray, cmap = 'gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#tao mot hinh ve moi
plt.figure()
plt.hist(im_array.flatten(), bins = 128)
plt.title("Biểu đồ Histogram của ảnh")
plt.xlabel("Giá trị của pixel")
plt.ylabel("Số lượng")
plt.show()
```

```python
# lam noi bat duong bien
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
